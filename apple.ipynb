```python
u_i=df["user_id"]
u_s=df["symbol_id"]
```

```python
df=pd.DataFrame()
```

```python
df["ui"]=u_i
df["us"]=u_s
df=df.sort_values(by='ui')
```

```python
i=0
p=0
last={}
while i <len(df)-1:
    if df.at[i,"ui"]==df.at[i+1,"ui"]:
        i=i+1
    else:
        si=[]
        temp=df.iloc[p:i+1,0:2]
        si=temp["us"]
        sii = list(set(si))
        last[df.at[i,"ui"]]=sii
        i=i+1
        p=i+1
```

```python
cf = pd.read_excel("/Users/lai/symbol_info.xlsx")
cf = cf.drop(["category","market","off_date"],axis = 1)
sym_id= []
for i in cf['symbol_id']:   
    sym_id.append(i)
```

```python
ind= []
for i in cf['industry']:   
    ind.append(i)
```

```python
new_Contrast={}
for i in range(len(sym_id)):
    
    new_Contrast[sym_id[i]]= ind[i]
```

```python
new_Contrast
```

```python
new_dict = {}
for key,value in last.items():
    tem = []
    for i in value:
        if i in new_Contrast:
            tem.append(new_Contrast[i])
    new_dict[key] = tem
```

```python
new_dict
```

```python
new_dict_est={}
for key in new_dict.keys():
    dict_num = {}
    x=new_dict[key]
    for j in x:
        if j in dict_num:
            dict_num[j]+=1
        else:
            dict_num[j]=1 
    for k,v in dict_num.items():
        if v==max(dict_num.values()):
            temp=k

    new_dict_est[key] = temp
    
```
