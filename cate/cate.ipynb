```python
cf1=cf.drop(["category","market","off_date"],axis = 1)
```

```python
list_=[]
for i in df["lists"]:
    list_.append(i)
```

```python
new_list=[]
for i in list_:
    temp=i.replace("'",'')
    new_list.append(temp)
```

```python
sym_id= []
for i in cf1['symbol_id']:   
    sym_id.append(i)
```

```python
ind= []
for i in cf1['industry']:   
    ind.append(i)
##類別
```

```python
new_Contrast={}
for i in range(len(sym_id)):
    
    new_Contrast[sym_id[i]]=[]
    new_Contrast[sym_id[i]].append(ind[i])
```

```python
cate=[]
for i in new_list: 
    k=i.replace("[",'')
    k=k.replace("]",'')
    k=k.replace(" ",'')
    q=k.split(",")
    temp=[]
    for j in q:
        name=new_Contrast.get(j)
        temp.append(name)
    cate.append(temp)
```
