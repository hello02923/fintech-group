```python
import csv
import os
import pandas as pd
path="/Users/lai/Desktop/20200504/history_data/2020-04-14_2020-04-23.csv"
##4/14-4/23瀏覽紀錄
```

```python
df=pd.read_csv(path)
df_user=df.drop_duplicates(subset="user_id")##去重複化id
```

```python
user_id=[]
for i in df_user['user_id']:
    user_id.append(i)
```

```python
df_sym=df.drop_duplicates(subset="symbol_id")
```

```python
sym_id= []
for i in df_sym['symbol_id']:   
    sym_id.append(i)
    
```

```python
uuser= []
for i in df_sym['user_id']:   
    uuser.append(i)
    
```

```python
user={}
for i in range(len(uuser)):
    
    if user.get(uuser[i]) == None:
        user[uuser[i]]=[]
        user[uuser[i]].append(sym_id[i])
    else:
        user[uuser[i]].append(sym_id[i])
    
```
